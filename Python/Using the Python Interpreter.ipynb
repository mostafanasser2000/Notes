Writing `python3` on the terminal will open an interactive shell which we can write python code there and run it

```python
!python3
```

Writing  `python3 -c statement` in the terminal will execute python statement

```python
!python3 -c "print('hello world')"
```

Write  `python3 -m <module name> [args]` in the terminal will execute module as python script

```python
!python3 -m rich
```
